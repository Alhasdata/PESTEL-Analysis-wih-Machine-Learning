```python
import pandas as pd
```

```python
dispo = pd.read_csv("DisponibiliteAlimentaire_2017.csv")
pop = pd.read_csv("Population_2000_2018.csv")
```

```python
dispo
```

```python
test = dispo[dispo["Produit"]=='Viande de Volailles']
test
```

```python
test2 = test[['Zone', 'Élément', 'Produit', 'Unité', 'Valeur',]]
```

```python
test2["Élément"] =test2["Élément"] + " (" + test2["Unité"]+")"
test2
```

```python
test3 = test2.pivot_table(index="Zone",columns=["Élément"])
test3.columns =["Alimentation pour touristes","Aliments pour animaux","Autres utilisations (non alimentaire)","Disponibilité alimentaire (Kcal/personne/jour)","Disponibilité alimentaire en quantité (kg/personne/an)","Disponibilité de matière grasse en quantité (g/personne/jour)","Disponibilité de protéines en quantité (g/personne/jour)","Disponibilité intérieure","Exportations - Quantité","Importations - Quantité","Nourriture","Pertes","Production","Résidus","Semences","Traitement","Variation de stock"]
```

```python
test3 = test3.fillna(0)
```

```python
test3["Autres utilisations"] = test3["Alimentation pour touristes"] + test3["Aliments pour animaux"] +test3["Autres utilisations (non alimentaire)"]+test3["Pertes"]+test3["Semences"]+test3["Traitement"]+test3["Résidus"]
```

```python
test3
```

```python
test4 =test3[["Disponibilité intérieure","Disponibilité alimentaire en quantité (kg/personne/an)","Importations - Quantité","Production","Exportations - Quantité","Variation de stock","Nourriture","Autres utilisations"]]
test4.columns=["Disponibilité intérieure (Milliers de tonnes)","Disponibilité alimentaire en quantité (kg/personne/an)","Importations - Quantité (Milliers de tonnes)","Production (Milliers de tonnes)","Exportations - Quantité (Milliers de tonnes)","Variation de stock (Milliers de tonnes)","Nourriture (Milliers de tonnes)","Autres utilisations (Milliers de tonnes)"]
```

```python
test4
```

```python
pop
```

```python
pop2 = pop[["Zone","Année","Valeur"]]
pop2 = pop2.pivot_table(index="Zone", columns="Année")
pop2
```

```python
pop2.columns = ["2000","2001","2002","2003","2004","2005","2006","2007","2008","2009","2010","2011","2012","2013","2014","2015","2016","2017","2018"]
```

```python
pop2["croissance_moyenne_population (%)"] = round(((pop2["2018"]/pop2["2013"])**(1/6)*100)-100,2)
```

```python
df = pd.merge(test4, pop2[["2017","croissance_moyenne_population (%)"]], on="Zone", how = "left")
df.rename(columns={"2017":"population_2017"})
df.reset_index()
```
